```
from perceptual_quality import nlpd
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
random_image_A = np.random.rand(256,256,3)
plt.figure()
plt.imshow(random_image_A)
plt.show()
```

```
random_image_B = np.random.rand(256,256,3)
plt.figure()
plt.imshow(random_image_B)
plt.show()
```

```
import tensorflow as tf
```

```
random_image_A_tf = tf.image.resize(random_image_A,(256,256))
random_image_A_tf = tf.expand_dims(random_image_A_tf, axis= 0)

random_image_B_tf = tf.image.resize(random_image_B,(256,256))
random_image_B_tf = tf.expand_dims(random_image_B_tf, axis= 0)
```

```
np.mean(nlpd.nlpd(random_image_A_tf, random_image_B_tf).numpy())
```
